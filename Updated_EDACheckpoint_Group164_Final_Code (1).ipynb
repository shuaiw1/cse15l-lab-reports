# Background and Prior Work

With the meteoric rise of Tesla Motors following the introduction of its luxury electric sports car, the Roadster, in 2008, the electric vehicle market gained unprecedented momentum, leading to accelerated innovation and competition across the automotive industry.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) Over the following years, other major automakers launched their own consumer-focused models like the Nissan LEAF in 2010 and the Chevrolet Bolt in 2017, which broadened accessibility and solidified EVs as a viable choice for the average consumer.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2)

Growing consumer interest in EVs has sparked extensive research and development in battery technology, driving down vehicle costs, enhancing performance, and making EVs more accessible and affordable.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) Additionally, governments around the world are bolstering support for robust EV charging networks through initiatives such as the U.S. National Electric Vehicle Infrastructure (NEVI) program and the European Union’s Alternative Fuels Infrastructure Regulation (AFIR), enabling greater access to reliable charging.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3)<sup>,</sup> <a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4) Furthermore, ambitious emissions reduction mandates and clean energy incentives, such as the American Federal Sustainability Plan, are reshaping the automotive industry and encouraging a continued transition to cleaner transportation methods.<a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5)

According to a 2023 study by the U.S. Bureau of Labor Statistics, consumer demand for EVs is projected to account for over 40 percent of total passenger car sales by 2030.<a name="cite_ref-6"></a>[<sup>6</sup>](#cite_note-6) The study found that environmental concerns are the primary driver of consumer adoption, with greater brand variety, increased vehicle ranges, and lower costs similarly contributing to sustained growth.<a name="cite_ref-6"></a>[<sup>6</sup>](#cite_note-6) Additionally, the study anticipates that tax incentives, rebates, and other government programs will continue to support adoption by offsetting high upfront vehicle costs.<a name="cite_ref-6"></a>[<sup>6</sup>](#cite_note-6) Based on these findings, our analysis aims to examine the impact of these technical and socioeconomic factors in driving EV adoption.

Additionally, a 2024 study by Pamidimukkala et al. analyzing contextual, situational, and psychological barriers and motivators associated with EV adoption identified air pollution reduction, supportive policies and incentives, accessible charging infrastructure, and vehicle range as the most influential factors driving consumer adoption.<a name="cite_ref-7"></a>[<sup>7</sup>](#cite_note-7) According to the findings, vehicle purchase price, driving range, and environmental considerations are the most significant factors influencing consumer adoption.<a name="cite_ref-7"></a>[<sup>7</sup>](#cite_note-7) Furthermore, accessible local and public charging infrastructure was found to play a crucial role in encouraging adoption across a broad demographic of consumers.<a name="cite_ref-7"></a>[<sup>7</sup>](#cite_note-7) We will incorporate these insights into our modeling approach to comprehensively assess the impact of these factors on EV adoption trends.

## Data Cleaning

### New ZEV Sales

To start, we load a CSV file containing the new ZEV sales data and display the DataFrame to examine the data's structure and contents.

```python
# Load a csv file of the new sales data
new_sales_df = pd.read_csv('./datasets/New_ZEV_Sales_Last_updated_08-06-2024_ada.xlsx - County.csv')

# Display the new sales data
new_sales_df
```

As shown, the data is already well-structured, with columns for year, quarter, county, fuel type, vehicle make and model, and sales figures.

Since the sales quarter column contains several missing values and is not relevant to our analysis, we’ll drop it to streamline the dataset.

```python
# Filter the sales data to only include figures from 2010 to 2023
new_sales_df = new_sales_df[(new_sales_df['Data_Year'] >= 2010) & (new_sales_df['Data_Year'] <= 2023)]

# Display the updated sales data
new_sales_df
```

Lastly, we’ll review the updated dataset to confirm the recent changes and verify its dimensions.

```python
# Display the first few rows of the cleaned data
new_sales_df.head()
```

```python
# Verify the dimensions of the cleaned data
print("Total elements:", new_sales_df.size, '\nTable shape:', new_sales_df.shape)
```

### Alternative Fueling Station Locations in California

With the sales and vehicle population data cleaned, our focus now shifts to preparing the Alternative Fueling Stations data for California.

As with the previous datasets, we’ll begin by loading the CSV file and displaying the DataFrame to review its structure and contents before proceeding with the cleaning process.

```python
# Load a csv file of fueling station locations data
fuel_station_df = pd.read_csv('./datasets/alt_fuel_stations (Nov 14 2024).csv', low_memory=False)

# Display the fuel station data
fuel_station_df
```

This dataset includes extensive information, such as geolocation data, operating hours, phone numbers, facility types, and more. Many of these columns are not essential for our analysis, so we’ll start by selecting only the relevant ones to condense the data and make it more manageable.

```python
# Specify the columns to include in the filtered DataFrame
columns_to_include = ['Fuel Type Code', 'ZIP', 'Open Date']

# Select the specified columns
fuel_station_df = fuel_station_df[columns_to_include]

# Display the DataFrame with the selected columns
fuel_station_df
```

With the dataset now condensed, we can begin cleaning the columns. We'll begin by checking for any missing values and removing them from the DataFrame to ensure data consistency.

```python
# Check for remaining missing values
fuel_station_df.isnull().sum(axis='rows')
```

```python
# Drop any rows with missing values
fuel_station_df = fuel_station_df.dropna()
```

Since many of the fuel types are non-zero-emission, we’ll filter the data to include only stations with the desired zero-emission fuel code.

```python
# Filter for stations with a zero-emission fuel type code
fuel_station_df = fuel_station_df[fuel_station_df['Fuel Type Code'] == 'ELEC']
```

We’ll also convert the 'Open Date' column to a date format and extract the year from each entry for further analysis.

```python
# Convert the Open Date column to datetime format and extract the year
fuel_station_df = fuel_station_df.copy()
fuel_station_df['Open Year'] = pd.to_datetime(fuel_station_df['Open Date']).dt.year

# Drop the original Open Date column
fuel_station_df = fuel_station_df.drop(columns=['Open Date'])
```

Notably, some of the open dates fall outside our analysis range, so we’ll filter out those years to ensure consistency.

```python
# Filter the fuel station data to only include figures from 2010 to 2023
fuel_station_df = fuel_station_df[(fuel_station_df['Open Year'] >= 2010) & (fuel_station_df['Open Year'] <= 2023)]
```

Since the previous datasets are already sorted chronologically, we’ll sort this dataset by Open Year as well.

```python
# Sort the DataFrame by chronological order
fuel_station_df = fuel_station_df.sort_values(by='Open Year')
```

Now, we’ll display the table to verify these changes before proceeding.

```python
# Display the first few rows of the updated data
fuel_station_df.head()
```

Next, we’ll standardize the location data by mapping each station’s ZIP code to its corresponding California county. This will align the location data with our other datasets, which use county-level information.

To begin, we’ll load the ZIP code data, filter for California records, and select only the ZIP and county_name columns.

```python
# Load a csv file of ZIP code data
zip_df = pd.read_csv('./datasets/uszips.csv')

# Select California from the ZIP code data 
zip_df = zip_df[zip_df['state_id'] == 'CA']

# Select only the ZIP code and county name columns
zip_df = zip_df[['zip', 'county_name']]

# Rename the zip column to align with the fuel station DataFrame
zip_df = zip_df.rename(columns={'zip': 'ZIP'})

# Display the ZIP code data for review
zip_df
```

Next, we’ll use the zip_df DataFrame to map county names onto the structures_df DataFrame based on matching ZIP codes.

```python
# Convert both ZIP code columns to string type
fuel_station_df['ZIP'] = fuel_station_df['ZIP'].astype(str)
zip_df['ZIP'] = zip_df['ZIP'].astype(str)

# Merge on the ZIP code column to add map county name to each station
fuel_station_df = fuel_station_df.merge(zip_df, on='ZIP', how='left')
```

After merging, we’ll remove the ZIP column, as it is no longer needed, and rename county_name to County for clarity.

```python
# Drop the ZIP code column
fuel_station_df = fuel_station_df.drop(columns=['ZIP'])

# Rename 'county_name' to 'County' for consistency
fuel_station_df = fuel_station_df.rename(columns={'county_name': 'County'})

# Display the first few rows of the updated data
fuel_station_df.head()
```

Finally, let’s check for and drop any remaining missing values and display the first few rows to review the changes.

```python
# Check for remaining missing values
fuel_station_df.isnull().sum(axis='rows')
```

```python
# Drop any rows with missing counties
fuel_station_df = fuel_station_df.dropna()

# Display the first few rows of the cleaned data
fuel_station_df.head()
```

```python
# Verify the dimensions of the cleaned data
print("Total elements:", fuel_station_df.size, '\nTable shape:', fuel_station_df.shape)
```

### Zero Emission Vehicle Laws and Incentives

```python
# Load a csv file of laws and incentives data
law_incentives_df = pd.read_csv('./datasets/laws_and_incentives (Nov 14 2024).csv')
```

As a last step,  let’s proceed with loading and cleaning the ZEV laws and incentives data.

```python
# Load a csv file of laws and incentives data
law_incentives_df = pd.read_csv('./datasets/laws_and_incentives (Nov 14 2024).csv')

# Display the laws and incentives data
law_incentives_df
```

This dataset contains extensive information, including policy titles and text, enacted dates, enacting agencies, amendments, and more. We’ll begin by selecting only the columns relevant to our analysis.

As our focus is on the effects of laws and incentives on EV adoption, we’re primarily interested in the type of incentive, key dates, and the technologies impacted.

```python
# Specify the columns to include in the filtered DataFrame
inc_columns_to_include = ['Law Id', 'Enacted Date', 'Type', 'Expired Date', 'Technology Categories', 'Incentive Categories', 'Status']

# Select the specified columns
law_incentives_df = law_incentives_df[inc_columns_to_include]

# Display the DataFrame with only the selected columns
law_incentives_df
```

With the relevant columns selected, let’s sort the dataset and standardize the dates by extracting the years to ensure consistency with our other datasets.

```python
# Convert the date columns to datetime format 
law_incentives_df = law_incentives_df.copy()
law_incentives_df['Enacted Date'] = pd.to_datetime(law_incentives_df['Enacted Date'])
law_incentives_df['Expired Date'] = pd.to_datetime(law_incentives_df['Expired Date'])

# Sort by enacted year in ascending order
law_incentives_df = law_incentives_df.sort_values(by='Enacted Date')

# Extract the years
law_incentives_df['Enacted Year'] = law_incentives_df['Enacted Date'].dt.year
law_incentives_df['Expired Year'] = law_incentives_df['Enacted Date'].dt.year

# Drop the original date columns
law_incentives_df = law_incentives_df.drop(columns=['Enacted Date', 'Expired Date'])
```

Next, we’ll verify that all laws and incentives fall within the specified date range for our analysis.

```python
# Filter rows to ensure enacted years fall within the specified range
law_incentives_df = law_incentives_df[(law_incentives_df['Enacted Year'] >= 2010) & (law_incentives_df['Enacted Year'] <= 2023)]
```

It's also important to note that some rows lack date information, making them challenging to incorporate into our analysis. Therefore, we’ll remove rows with missing dates.

```python
# Drop rows where all date columns are missing
law_incentives_df = law_incentives_df.dropna(subset=['Enacted Year', 'Expired Year'], how='all')
```

As a final step, we’ll inspect the updated dataset to verify the applied changes and confirm its dimensions.

```python
# Display the first few rows of the cleaned data
law_incentives_df.head()
```

```python
# Verify the dimensions of the cleaned data
print("Total elements:", law_incentives_df.size, '\nTable shape:', law_incentives_df.shape)
```

### Examining ZEV Sales and Vehicle Population by County

To begin our exploratory data analysis, we examine the new ZEV sales and vehicle population datasets to identify trends and patterns that provide insights into ZEV adoption across California from 2010 to 2023.

We will first review the descriptive statistics for the new ZEV sales dataset to understand the distribution of sales across counties and identify any potential outliers. Since the dataset is broken down by year, we will start by aggregating the data to present a comprehensive overview of ZEV sales over this period.

```python
# Aggregate ZEV sales by county
county_sales_df = new_sales_df.groupby('County')['Number of Vehicles'].sum().reset_index()

# Display the descriptive statistics
county_sales_df.describe()
```

As shown, there is significant variability in ZEV sales across counties during this time. The total number of vehicles sold ranges from a minimum of 8 to a maximum of 482,178, highlighting a substantial disparity in adoption rates across counties. 

To delve deeper into this disparity, we will identify the counties with the highest and lowest ZEV sales.

```python
# Find the county with the highest sales
highest_sales = county_sales_df.nlargest(10, 'Number of Vehicles')

# Find the county with the lowest sales
lowest_sales = county_sales_df.nsmallest(10, 'Number of Vehicles')

# Display the county lists
print('Counties with the highest sales: ')
print(highest_sales)

print('\nCounties with the lowest sales: ')
print(lowest_sales)
```

As expected, the counties with the highest ZEV sales are those with large populations and well-developed urban infrastructure, while the counties with the lowest sales are smaller, more rural regions with likely limited infrastructure and lower population densities.

To effectively visualize this variability, we will utilize a combination of bar plots and geographic maps.

```python
# Sort data for improved flow in visualization
county_sales_df = county_sales_df.sort_values(by='Number of Vehicles', ascending=False)

# Plot the county sales data
plt.figure(figsize=(10, 6))
sns.barplot(
    data=county_sales_df,
    x='County',
    y='Number of Vehicles',
    palette='Greens_r',
    hue='County',
    legend=False
)

plt.title('ZEV Sales in California by County (2010-2023)', fontsize=16)
plt.xlabel('County', fontsize=14)
plt.ylabel('Number of Vehicles Sold', fontsize=14)

# Rotate county names for readability
plt.xticks(rotation=90, fontsize=10)
plt.show()
```

```python
# Load geospatial data for California counties
county_shapes_gdf = gpd.read_file('./shapefiles/CA_Counties.shp')

# Merge geographic data with county sales data
county_shapes_gdf = county_shapes_gdf.rename(columns={'NAME': 'County'})
sales_map_data = county_shapes_gdf.merge(county_sales_df, on='County', how='left')

# Plot the county sales map
fig, ax = plt.subplots(figsize=(10, 10))
sales_map_data.plot(column='Number of Vehicles', 
    cmap='Greens', 
    legend=True, 
    ax=ax, 
    edgecolor='grey',
)

# Adjust tile and axis configuration
ax.set_title('ZEV Sales in California by County (2010-2023)')
plt.axis('off')
plt.show()
```

Now, we will perform a similar analysis for the total ZEV population during the same period to identify patterns and disparities across counties.

```python
# Aggregate ZEV population by county
county_vehicle_df = vehicle_pop_df.groupby('County')['Number of Vehicles'].sum().reset_index()

# Find the county with the highest ZEV population
highest_pop = county_vehicle_df.nlargest(10, 'Number of Vehicles')

# Find the county with the lowest ZEV population
lowest_pop = county_vehicle_df.nsmallest(10, 'Number of Vehicles')

# Display the county lists
print('Counties with the highest ZEV populations: ')
print(highest_pop)

print('\nCounties with the lowest ZEV populations: ')
print(lowest_pop)
```

### Charging Station Availability by County 

We want to examine the charging station dataset in order to see the influence that accessibility to chargers has on adoption rates of ZEVs across California. 

In order to visualize this data, we will create a barplot and a geographical map that depicts the disparities in charging stations by county. First, we will aggregate by county to examine county-level differences in charging infrustructure availability.

```python
# Aggregate Charging Stations by County
county_stations_df = fuel_station_df.groupby("County").size().reset_index(name='Count')

most_stations = county_stations_df.nlargest(10, 'Count')
least_stations = county_stations_df.nsmallest(10, 'Count')

# Display the county lists
print('Counties with the most ZEV fueling stations: ')
print(most_stations)

print('\nCounties with the least ZEV fueling stations: ')
print(least_stations)
```

Unsurprisingly, the counties with the most available charging infrastructure closely align with those that lead in ZEV sales and total vehicle population. On the other hand, counties with the least fueling stations also exhibited some of the lowest ZEV sales and population.
