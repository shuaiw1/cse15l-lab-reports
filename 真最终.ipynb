# Abstract

The transition to zero-emission vehicles (ZEVs) comprises a critical component of California's strategy to combat climate change and meet greenhouse gas reduction targets. The state has led the way in promoting ZEV adoption through a combination of investments in technology, progressive policies and financial incentives, and development of robust charging infrastructure. Despite the substantial progress, understanding the specific factors that drive significant ZEV adoption remains essential for refining future strategies and replicating success in other states. Prior studies have highlighted the significance of incentives and technology advancements in promoting electric vehicle adoption, but few have comprehensively examined the collective impact of these factors over an extended period. This analysis aims to bridge that gap by analyzing the relationships between advancements in ZEV technology, charing infrastructure availability, and regulatory actions and financial incentives in driving ZEV adoption in California from 2010 to 2023. By doing so, it aims to identify key drivers of growth and provide insights into future adoption trends. 

Using a combination of sales data, legislative records, and infrastructure metrics, this analysis examines trends in ZEV sales growth alongside significant advancements in ZEV technology, major policy milestones, and significant infrastructure development projects. Additionally, this analysis employs visualizations such as scatter plots, bar charts, and line plots to highlight key correlations and patterns over time. The study identifies strong correlations between policy enactments and surges in ZEV sales, with key milestones like the Clean Vehicle Rebate Project (CVRP) and the 2035 ZEV Mandate playing pivotal roles. The findings also reveal compounding effects of sustained incentives and infrastructure development on adoption rates. These results emphasize the importance of comprehensive and sustained approach to supporting the transition to ZEVs.

### Alternative Fueling Station Locations in California

With the sales and vehicle population data cleaned, our focus now shifts to preparing the Alternative Fueling Stations data for California.

As with the previous datasets, we’ll begin by loading the CSV file and displaying the DataFrame to review its structure and contents before proceeding with the cleaning process.

```python
# Load a csv file of fueling station locations data
fuel_station_df = pd.read_csv('./datasets/alt_fuel_stations (Nov 14 2024).csv', low_memory=False)

# Display the fuel station data
fuel_station_df
```

This dataset includes extensive information, such as geolocation data, operating hours, phone numbers, facility types, and more. Many of these columns are not essential for our analysis, so we’ll start by selecting only the relevant ones to condense the data and make it more manageable.

```python
# Specify the columns to include in the filtered DataFrame
columns_to_include = ['Fuel Type Code', 'ZIP', 'Open Date']

# Select the specified columns
fuel_station_df = fuel_station_df[columns_to_include]

# Display the DataFrame with the selected columns
fuel_station_df
```

With the dataset now condensed, we can begin cleaning the columns. We'll begin by checking for any missing values and removing them from the DataFrame to ensure data consistency.

```python
# Check for remaining missing values
fuel_station_df.isnull().sum(axis='rows')
```

```python
# Drop any rows with missing values
fuel_station_df = fuel_station_df.dropna()
```

Since many of the fuel types are non-zero-emission, we’ll filter the data to include only stations with the desired zero-emission fuel code.

```python
# Filter for stations with a zero-emission fuel type code
fuel_station_df = fuel_station_df[fuel_station_df['Fuel Type Code'] == 'ELEC']
```

We’ll also convert the 'Open Date' column to a date format and extract the year from each entry for further analysis.

```python
# Convert the Open Date column to datetime format and extract the year
fuel_station_df = fuel_station_df.copy()
fuel_station_df['Open Year'] = pd.to_datetime(fuel_station_df['Open Date']).dt.year

# Drop the original Open Date column
fuel_station_df = fuel_station_df.drop(columns=['Open Date'])
```

Notably, some of the open dates fall outside our analysis range, so we’ll filter out those years to ensure consistency.

```python
# Filter the fuel station data to only include figures from 2010 to 2023
fuel_station_df = fuel_station_df[(fuel_station_df['Open Year'] >= 2010) & (fuel_station_df['Open Year'] <= 2023)]
```

Since the previous datasets are already sorted chronologically, we’ll sort this dataset by Open Year as well.

```python
# Sort the DataFrame by chronological order
fuel_station_df = fuel_station_df.sort_values(by='Open Year')
```

Now, we’ll display the table to verify these changes before proceeding.

```python
# Display the first few rows of the updated data
fuel_station_df.head()
```

Next, we’ll standardize the location data by mapping each station’s ZIP code to its corresponding California county. This will align the location data with our other datasets, which use county-level information.

To begin, we’ll load the ZIP code data, filter for California records, and select only the ZIP and county_name columns.

```python
# Load a csv file of ZIP code data
zip_df = pd.read_csv('./datasets/uszips.csv')

# Select California from the ZIP code data 
zip_df = zip_df[zip_df['state_id'] == 'CA']

# Select only the ZIP code and county name columns
zip_df = zip_df[['zip', 'county_name']]

# Rename the zip column to align with the fuel station DataFrame
zip_df = zip_df.rename(columns={'zip': 'ZIP'})

# Display the ZIP code data for review
zip_df
```

Next, we’ll use the zip_df DataFrame to map county names onto the structures_df DataFrame based on matching ZIP codes.

```python
# Convert both ZIP code columns to string type
fuel_station_df['ZIP'] = fuel_station_df['ZIP'].astype(str)
zip_df['ZIP'] = zip_df['ZIP'].astype(str)

# Merge on the ZIP code column to add map county name to each station
fuel_station_df = fuel_station_df.merge(zip_df, on='ZIP', how='left')
```

After merging, we’ll remove the ZIP column, as it is no longer needed, and rename county_name to County for clarity.

```python
# Drop the ZIP code column
fuel_station_df = fuel_station_df.drop(columns=['ZIP'])

# Rename 'county_name' to 'County' for consistency
fuel_station_df = fuel_station_df.rename(columns={'county_name': 'County'})

# Display the first few rows of the updated data
fuel_station_df.head()
```

Finally, let’s check for and drop any remaining missing values and display the first few rows to review the changes.

```python
# Check for remaining missing values
fuel_station_df.isnull().sum(axis='rows')
```

```python
# Drop any rows with missing counties
fuel_station_df = fuel_station_df.dropna()

# Display the first few rows of the cleaned data
fuel_station_df.head()
```

### Examining ZEV Sales and Vehicle Population by County

To begin our exploratory data analysis, we examine the new ZEV sales and vehicle population datasets to identify trends and patterns that provide insights into ZEV adoption across California from 2010 to 2023.

We will first review the descriptive statistics for the new ZEV sales dataset to understand the distribution of sales across counties and identify any potential outliers. Since the dataset is broken down by year, we will start by aggregating the data to present a comprehensive overview of ZEV sales over this period.

```python
# Aggregate ZEV sales by county
county_sales_df = new_sales_df.groupby('County')['Number of Vehicles'].sum().reset_index()

# Display the descriptive statistics
county_sales_df.describe()
```

As shown, there is significant variability in ZEV sales across counties during this time. The total number of vehicles sold ranges from a minimum of 8 to a maximum of 482,178, highlighting a substantial disparity in adoption rates across counties. 

To delve deeper into this disparity, we will identify the counties with the highest and lowest ZEV sales.

```python
# Find the county with the highest sales
highest_sales = county_sales_df.nlargest(10, 'Number of Vehicles')

# Find the county with the lowest sales
lowest_sales = county_sales_df.nsmallest(10, 'Number of Vehicles')

# Display the county lists
print('Counties with the highest sales: ')
print(highest_sales)

print('\nCounties with the lowest sales: ')
print(lowest_sales)
```

As expected, the counties with the highest ZEV sales are those with large populations and well-developed urban infrastructure, while the counties with the lowest sales are smaller, more rural regions with likely limited infrastructure and lower population densities.

To effectively visualize this variability, we will utilize a combination of bar plots and geographic maps.

```python
# Sort data for improved flow in visualization
county_sales_df = county_sales_df.sort_values(by='Number of Vehicles', ascending=False)

# Plot the county sales data
plt.figure(figsize=(20, 10))
sns.barplot(
    data=county_sales_df,
    x='County',
    y='Number of Vehicles',
    palette='Greens_r',
    hue='County',
    legend=False
)

plt.title('ZEV Sales in California by County (2010-2023)')
plt.xlabel('County')
plt.ylabel('Number of Vehicles Sold')

# Rotate county names for readability
plt.xticks(rotation=90)
plt.show()
```

```python
# Load geospatial data for California counties
county_shapes_gdf = gpd.read_file('./shapefiles/CA_Counties.shp')

# Merge geographic data with county sales data
county_shapes_gdf = county_shapes_gdf.rename(columns={'NAME': 'County'})
sales_map_data = county_shapes_gdf.merge(county_sales_df, on='County', how='left')

# Plot the county sales map
fig, ax = plt.subplots(figsize=(10, 10))
sales_map_data.plot(column='Number of Vehicles', 
    cmap='Greens', 
    legend=True, 
    ax=ax, 
    edgecolor='grey',
)

# Adjust tile and axis configuration
ax.set_title('ZEV Sales in California by County (2010-2023)')
plt.axis('off')
plt.show()
```

Now, we will perform a similar analysis for the total ZEV population during the same period to identify patterns and disparities across counties.

```python
# Aggregate ZEV population by county
county_vehicle_df = vehicle_pop_df.groupby('County')['Number of Vehicles'].sum().reset_index()

# Find the county with the highest ZEV population
highest_pop = county_vehicle_df.nlargest(10, 'Number of Vehicles')

# Find the county with the lowest ZEV population
lowest_pop = county_vehicle_df.nsmallest(10, 'Number of Vehicles')

# Display the county lists
print('Counties with the highest ZEV populations: ')
print(highest_pop)

print('\nCounties with the lowest ZEV populations: ')
print(lowest_pop)
```

This analysis reveals consistent patterns in the counties with the highest and lowest ZEV adoption levels. Los Angeles, Orange, and Santa Clara counties lead in both total sales and vehicle population, reflecting strong support for ZEV adoption. Conversely, Modoc, Sierra, and Alpine counties consistently report the lowest numbers in both sales and population, underscoring the challenges faced by rural areas in adopting ZEVs. 

As before, we will use a combination of bar plots and geospatial maps to better visualize these disparities.

```python
# Sort data for improved flow in visualization
county_vehicle_df = county_vehicle_df.sort_values(by='Number of Vehicles', ascending=False)

# Plot the county sales data
plt.figure(figsize=(20, 10))
sns.barplot(
    data=county_vehicle_df,
    x='County',
    y='Number of Vehicles',
    palette='Greens_r',
    hue='County',
    legend=False
)

plt.title('ZEV Population in California by County (2010-2023)')
plt.xlabel('County')
plt.ylabel('Number of Vehicles')

# Rotate county names for readability
plt.xticks(rotation=90)
plt.show()
```

```python
# Merge geographic data with county sales data
vehicle_map_data = county_shapes_gdf.merge(county_vehicle_df, on='County', how='left')

# Plot the county sales map
fig, ax = plt.subplots(figsize=(10, 10))
vehicle_map_data.plot(column='Number of Vehicles', 
    cmap='Greens', 
    legend=True, 
    ax=ax, 
    edgecolor='grey',
)

# Adjust tile and axis configuration
ax.set_title('ZEV Population in California by County (2010-2023)')
plt.axis('off')
plt.show()
```

Unsurprisingly, the counties that we had identified as having the most EV sales also exhibit the highest ZEV populations in the state, which suggests a direct relationship between sustained sales activity and the accumulation of ZEV ownership over time.

```python
import statsmodels.api as sm
from patsy import dmatrices
combined_df = county_sales_df.merge(
    county_stations_df.rename(columns={'Count': 'Station Count'}),
    on='County',
    how='left'
)
combined_df['Growth Rate'] = combined_df['Number of Vehicles'] / combined_df['Station Count']
combined_df = combined_df.dropna()
# Redefine column names, removing spaces for subsequent analysis
combined_df = combined_df.rename(columns={'Growth Rate': 'GrowthRate', 'Station Count': 'StationCount'})

# Define the formula
formula = 'GrowthRate ~ StationCount'
# Build dmatrices
y, X = dmatrices(formula, data=combined_df, return_type='dataframe')

# OLS modeling and fitting
model = sm.OLS(y, X)
result = model.fit()

print(result.summary())
```

From this result we can see that the lowest counties have very poor infrastructure, with Modoc, Sierra, and Alpine having extremely low numbers of charging stations and should prioritize increased infrastructure development. Local governments should likely provide policy incentives to support local charging station construction. For counties with high growth rates but insufficient charging stations, optimize the layout of existing charging stations to improve the user experience.

```python
# Extract t-test values
t_test_values = result.tvalues
p_values = result.pvalues

# Print t-test and p-values
print("\nT-test values:")
print(t_test_values)

print("\nP-values:")
print(p_values)

# Converted to DataFrame format for easy visualization
t_test_df = pd.DataFrame({
    'Variable': t_test_values.index,
    'T-test value': t_test_values.values,
    'P-value': p_values.values
})

# Display the T-test form
print("\nT-test Results:")
print(t_test_df)

# 可视化T-test值
plt.figure(figsize=(8, 5))
sns.barplot(data=t_test_df, x='Variable', y='T-test value')
plt.axhline(0, color='gray', linestyle='--')
plt.title('T-test Values for Predictors')
plt.xlabel('Predictor Variables')
plt.ylabel('T-test Value')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

From the previous arguments and the T-test values here, we can see that the number of charging stations has a significant positive effect on the growth rate, so for low-will counties, growing charging stations is very important.

# Discusison and Conclusion

California's transition to zero-emission vehicles(ZEVs) provides valuable insights as to how policies, infrastructure and technology shape the adoption of sustainable transportation. By examining data on ZEV sales, vehicle populations and charging infrastructure across California's counties from 2010-2023, we see patterns that highlight what drives and what refutes ZEV adoption. These patterns are highlighted in a variety of ways: bar plots, geographical maps, line plots and scatter plots.

When looking at ZEV sales by county, we see a significant disparity in adoption rates across California. Los Angeles, Orange and Santa Clara counties dominate these sales(Los Angeles accounts for 482,000+ vehicles alone!) during this period. In contrast, rural areas like Modoc and Sierra reported minimal sales with less than 20 ZEV sales. This analysis usggests that urban areas are more conducive to ZEV adoption. The distribution of ZEV population closesly mirrors these trends emphasizing the effect of sales on vehicle population over time. The correlation found suggests regions with early adoption of ZEVs continue to lead the market, most likely as a result of established infrastructure and community norms.

In terms of trend sales over time, we see the gradual adoption of ZEVs in 2010, followed by an exponential rise in 2020. The early years may reflect the challenges of marketing, high costs, and limited model availability. However, the rise in adoption reflects technological advancements that made ZEVs more accessible. The post-2020 surge reflects heightened customer awareness, stronger incentives and ambitious regulatory targets.

Charging station availability plays a crucial role in ZEV adoptions. Counties with a plethora of accessible charging infrastructures, such as Los Angeles and Orange counties, lead in both ZEV sales and population. While Plumas and Colusa lag behind due to their sparce infrastructures. The increase in charging stations from 2010-2018 was steady and followed by a sharp rise in numbers after 2019. This spike conflates with the sales and poulation as the demand for them was higher. The creation of new stations then came to a steady flow in 2023, a sign that there was either market saturation or slower expansion in underdeveloped areas.

Looking at model popularity, it is evident that Yesla is a dominant company in the California ZEV market. With their 2170 battery cells and structural battery packs, it is no wonder that there was a significant increase in sales. Notably, Tesla Model 3 and Model Y as they account for a large number of sales, underscoring strong brand preference and consumer confidence in these models.

As for regulatory measures and incentives, they have shaped California's ZEV landscape. Policies like the Clean Vehicle Rebate Project(CVRP) and mandates like the Advances Clean Trucks Regulation have increased adoption by reducing financial barriers and fostering market confidence. We see peaks in 2012, 2018, and 2023, which all align with legislative actions. While we must consider all factors, there is no doubt that the consistent rise in adoption rates can also be attributed to legislative efforts.

Despite theses successes, some challenges stil remain. The overrepresentation of ZEV adoption in urban areas highlight the disparities in infratructure and policies in rural regions. This imbalance can skew the perceptions of the statewide adoption and hide barriers that these under-resources communities face. Additionally, while our anaysis draw from publicly available information, there are still ethical concerns around representing these trends in a fair and unbiased way.

Overall, California's experience with ZEV adoption demonstrates the power that infrastructure, policy and technology all have when working together. Urban centers lead the transition into an environmentally favorable lifestyle, yet we stll need to address disparities to achieve equitable and widespread ZEV adoption. Examining these trends and what drives them, stakeholders can anaylze and change things in order to create a smoother transition towards a zero-emission future.
